```python
#dropping_tables
drop=df.iloc[:,19:31]
df=df.drop(drop,axis=1)
df=df.drop(['WindGustSpd','PoorWeather'],axis=1)
```

```python
#finding_missing_values_using_missingno
import missingno as msno
msno.bar(df)
plt.show()
```

```python
#filling_missing_values_using_imputer
from sklearn.impute import SimpleImputer
impute=SimpleImputer(missing_values=np.nan,strategy='mean')
for col in df.columns:
  df.iloc[:,11:16]=impute.fit_transform(df.iloc[:,11:16])
```

```python
impute=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
for col in df.columns:
  df.iloc[:,6:17]=impute.fit_transform(df.iloc[:,6:17])
```

```python
#Label_Encoding
from sklearn.preprocessing import LabelEncoder
a=LabelEncoder()
cat=df.select_dtypes(include=['object']).columns
for col in cat:
  df[col] = df[col].astype(str)
  df[col]=a.fit_transform(df[col])
```

```python
#heatmap
plt.figure(figsize=(20,15))
corr=df.corr()
sns.heatmap(corr,annot=True,fmt='.2g',linewidth=1)
plt.show()
```

```python
x=df.drop('MaxTemp',axis=1)
y=df['MaxTemp']
```

```python
#Linear_Regression
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x,y)
```

```python
#accuracy_score
reg.score(x,y)
```

```python
#coefficient
reg.coef_
```

```python
#intercept
reg.intercept_
```

```python
#training_and_testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
x_train
```

```python
reg.fit(x_train,y_train)
```

```python
y_pred1=reg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,r2_score,mean_absolute_error,mean_squared_error
#r2_score
r2_score(y_test,y_pred1)
```

```python
#mean_absolute_error
mean_absolute_error(y_test,y_pred1)
```

```python
#mean_squared_error
mean_squared_error(y_test,y_pred1)
```

```python
#root_mean_squared_error
np.sqrt(mean_squared_error(y_test,y_pred1))
```

```python
y_pred2=reg.predict(x_train)
```
